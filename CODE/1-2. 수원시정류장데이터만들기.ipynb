```python
import pandas as pd
juso_df=pd.read_csv('/content/gdrive/MyDrive/★데이터분석언어/2022년_전국버스정류장 위치정보_데이터.csv', encoding='utf-8')
juso_df.head(3)
```

```python
juso_df=juso_df.loc[juso_df['도시명']=='수원시']
juso_df.head(3)
```

```python
juso_df=juso_df[['정류장명', '위도', '경도', '모바일단축번호','도시명', '관리도시명']]
```

```python
# 주소 코드 찾기
from geopy.geocoders import Nominatim

def geocoding_reverse(lat_lng_str):
    geolocoder = Nominatim(user_agent = 'South Korea', timeout=None)
    address = geolocoder.reverse(lat_lng_str)
    return address
```

```python
juso_df['위도']=juso_df['위도'].astype('str')
juso_df['경도']=juso_df['경도'].astype('str')
juso_df['lat_lng']=juso_df['위도']+", "+juso_df['경도']
juso_df['lat_lng']=juso_df['lat_lng'].astype('str')
```

```python
address_result=[]
for i in juso_df['lat_lng']:
  try:
    a=geocoding_reverse(i)
    address_result.append(a)
  except:
    address_result.append('recheck')
```

```python
juso_df2=pd.DataFrame(address_result)
```

```python
jusodf1=juso_df.reset_index()
```

```python
final_juso=pd.concat([jusodf1, juso_df2], axis=1)
final_juso=final_juso.rename(columns={0 : '주소'})
```

```python
final_juso.head(3)
```

```python
final_juso=final_juso[['모바일단축번호', '정류장명', '위도', '경도', '주소']]
```
