# 광역버스 정류장 위치 시각화

```python
import folium #지도 시각화 라이브러리
df_busstop=df.drop_duplicates(['모바일단축번호'], ignore_index=True) #중복된 항목 제거
df_busstop=df_busstop[['정류장', '위도', '경도', '모바일단축번호', '정류장_시', '정류장_구']]
df_busstop_suwon=df_busstop.loc[df_busstop['정류장_시']=='수원시']
df_busstop_suwon.isnull().sum() #결측값 체크
df_busstop_suwon=df_busstop_suwon[['정류장', '위도', '경도', '정류장_구']].dropna() #결측값 제거
df_busstop_suwon=df_busstop_suwon.reset_index()
```

```python
df=df_busstop_suwon

from folium.plugins import MiniMap, MarkerCluster
center=[37.2635727,127.0286009]
m=folium.Map(location=center, zoom_start=13)
c_dic={'장안구' : 'red', '팔달구' : 'orange', '권선구' : 'green', '영통구' : 'blue'} #해당 정류장을 구별로 다른 색으로 표시함
#클러스터 추가하기
marker_cluster = MarkerCluster().add_to(m)
for i in df.index:
  folium.Marker(location=list(df.loc[i, ['위도', '경도']]), popup=df.loc[i,'정류장'], icon=folium.Icon(color=c_dic[df.loc[i, '정류장_구']], icon='star')).add_to(m)
```
