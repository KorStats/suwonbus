# 시간대별 분석

```python
df1=pd.read_csv('/content/gdrive/MyDrive/★데이터분석언어/분석과제/이슈분석/★시계열_23년 평일_연휴제거.csv', encoding='utf-8-sig')
df2=pd.read_csv('/content/gdrive/MyDrive/★데이터분석언어/분석과제/이슈분석/★시계열_22년 평일_연휴제거.csv', encoding='utf-8-sig')
```

```python
df1=df1.drop(['일자', '정류장순번', '정류장', '위도', '경도', '모바일단축번호', '정류장_시', '정류장_구', '도시명', '관리도시명', 'total', 'total_1', '요일'], axis=1)
group1 = df1.groupby(['노선']).sum()
total_sum_1= group1.sum()
df2=df2.drop(['일자', '정류장순번', '정류장', '위도', '경도', '모바일단축번호', '정류장_시', '정류장_구', '도시명', '관리도시명', 'total', 'total_1', '요일'], axis=1)
group2 = df2.groupby(['노선']).sum()
total_sum_2= group2.sum()
```

```python
result=pd.DataFrame(total_sum_1, columns=["2023"])
result['2022']=total_sum_2.values
```

```python
import matplotlib.pyplot as plt
plt.rc('font', family='NanumBarunGothic')
plt.plot(result.index, result['2022'], label='2022년')
plt.plot(result.index, result['2023'], label='2023년')
plt.legend()
plt.xlabel('시간대')
plt.ylabel('승차인원 합계')
plt.title('시간대별 승차인원 합계')
plt.show()
```

# 노선 분석

```python
import pandas as pd
df1=pd.read_csv('/content/gdrive/MyDrive/★데이터분석언어/분석과제/이슈분석/★시계열_22년 평일_연휴제거.csv', encoding='utf-8-sig')
df2=pd.read_csv('/content/gdrive/MyDrive/★데이터분석언어/분석과제/이슈분석/★시계열_23년 평일_연휴제거.csv', encoding='utf-8-sig')
```

```python
df1['total_1']=df1['6']+df1['7']+df1['8'] #출근시간 승차인원만 저장
df2['total_1']=df2['6']+df2['7']+df2['8']
```

```python
cnt1=df1.groupby(['노선','모바일단축번호', '정류장', '정류장순번'])['total_1'].sum() #22년 평일 출근시간대 승차인원 수 합
cnt2=df2.groupby(['노선','모바일단축번호', '정류장','정류장순번'])['total_1'].sum() #23년 평일 출근시간대 승차인원 수 합
```

```python
df_22=pd.DataFrame(cnt1).reset_index()
df_23=pd.DataFrame(cnt2).reset_index()
```

```python
df_23=df_23.sort_values(['노선', '정류장순번'])
df_22=df_22.sort_values(['노선', '정류장순번'])
```

```python
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
result = pd.merge(df_23,df_22, how='left',on=['노선', '정류장'])
```

```python
result.sort_values(['노선', '정류장순번_x'])
result=result[['노선', '정류장', '정류장순번_x','total_1_x', 'total_1_y']]
result['gap']=result['total_1_x']-result['total_1_y'] #23년-22년 : gap
```

# 승차 gap 분석(출근시간대)

```python
import pandas as pd
df1=pd.read_csv('/content/gdrive/MyDrive/★데이터분석언어/분석과제/이슈분석/22년도 출근시간대 버스승차.csv', encoding='utf-8-sig')
df2=pd.read_csv('/content/gdrive/MyDrive/★데이터분석언어/분석과제/이슈분석/23년도 출근시간대 버스승차.csv', encoding='utf-8-sig')
```

```python
df_22=df1.sort_values(['노선', '정류장순번'])
df_23=df2.sort_values(['노선', '정류장순번'])
```

```python
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
result = pd.merge(df_23,df_22, how='left',on=['노선', '정류장', '정류장순번'])
result=result[['노선', '정류장', '정류장순번','total_1_x', 'total_1_y']]
```

```python
result['gap']=result['total_1_x']-result['total_1_y'] #23년-22년 : gap
```

```python
df1['total_1']=df1['6']+df1['7']+df1['8']
df2['total_1']=df2['6']+df2['7']+df2['8']
```

```python
cnt1=df1.groupby(['노선','모바일단축번호', '정류장', '정류장순번'])['total_1'].sum() #23년 평일 출근시간대 승차인원 수 합
cnt2=df2.groupby(['노선','모바일단축번호', '정류장','정류장순번'])['total_1'].sum() #22년 평일 출근시간대 승차인원 수 합
```

```python
result1=pd.DataFrame(cnt1).reset_index()
result2=pd.DataFrame(cnt2).reset_index()
result1=result1.set_index(['노선', '모바일단축번호'])
result2=result2.set_index(['노선', '모바일단축번호'])
```

```python
result = pd.merge(result1,result2, how='outer',on=['노선', '모바일단축번호'])
```
